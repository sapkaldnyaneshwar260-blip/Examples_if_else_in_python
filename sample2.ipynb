```python
a = int(input("Enter the Number:"))
s=0

while a!=0:
        rem = a%10
        if rem==0:
           a = a//10 
           continue

        else:
            s= s*10+rem 
        a = a//10 
        
print(s)
rev =0
#reverse
while s!=0:
     r = s%10
     rev = rev*10+r
     s = s//10
print("reverse = ",rev)     
     
```

```python
a = int(input("Enter the Number:"))
b =5000
small=0
while a!=0:
    rem = a%10

    if rem<b:
        small=rem 
        b=rem

    a = a//10  
print("Smallest = ",small)     
```
