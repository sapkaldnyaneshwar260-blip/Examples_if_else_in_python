```python
x = int(input("Enter the x co-ordinate:"))
y = int(input("Enter the y co-ordinate:"))
print(f"x ={x} and y ={y}") 
if(x==0 and y==0):
    print("Origin")
elif(x ==0 and y!=0 ):
    print("On y-axis")
elif(x!=0 and y==0):
    print("On x-axis ")
elif(x>0 and y>0):
    print("First Quadrant(I)")
elif(x>0 and y<0):
    print("Fourth Quadrant(IV)")
elif(x<0 and y<0):
    print("Third Quadrant(III)")   
elif(x<0 and y>0):
    print("Second Quadrant(II)")
```
